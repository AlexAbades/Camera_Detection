```python
def segmentation_histogram(ax, I, S, edges=None):
    '''
    Histogram for data and each segmentation label.
    '''
    if edges is None:
        edges = np.linspace(I.min(), I.max(), 100)
    ax.hist(I.ravel(), bins=edges, color = 'k')
    centers = 0.5*(edges[:-1] + edges[1:]);
    for k in range(S.max()+1):
        ax.plot(centers, np.histogram(I[S==k].ravel(), edges)[0])
```

```python
def histo(im):

    s = im.shape

    if len(s) == 2:
        n, bins, patches = plt.hist(x=im.ravel(), 
                                    bins=50, 
                                    alpha=0.7, 
                                    rwidth=0.85)
        plt.grid(axis='y', alpha=0.75)
        plt.xlabel('Intensity')
        plt.ylabel('Frequency')
        plt.title('Image Intensities')
        plt.xlim((0,255))
    
    else: 
        color = ['r', 'g', 'b']
        for chanel, c in zip(range(len(s)), color):
            n, bins, patches = plt.hist(x=im[:,:,chanel].ravel(), 
                                    bins=50, 
                                    color= c,
                                    alpha=0.7, 
                                    rwidth=0.85)
        plt.grid(axis='y', alpha=0.75)
        plt.xlabel('Intensity')
        plt.ylabel('Frequency')
        plt.title('Image Intensities')
        plt.xlim((0,255))


histo(gray)
histo(im1)
```
